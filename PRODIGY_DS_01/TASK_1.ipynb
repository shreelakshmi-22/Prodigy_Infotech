### EXPLORATORY DATA ANALYSIS

```python
plt.figure(figsize=(25,10))
sns.countplot(data=data,x='Age',hue='HeartDisease')
```

* From the above barplot we can see that the rate of heart disease is higher from the age of 50 to 65

```python
plt.figure(figsize=(5,3))
sns.countplot(data=data,x='Sex')
```

* This plot shows the distribution of Gender across the entire dataset.

```python
sns.histplot(data=data,x='Age',kde=True)
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
